# Data Dictionary

### Admin
- Repräsentiert den Spielleiter
- Funktionen zur Verwaltung des Spiels: Starten, Pausieren, Fortsetzen, Beenden
- Überwacht Spielfortschritt und Regelkonformität

### Aktionskarte
- Erlaubt besondere Aktionen (z.B. Räuber versetzen)
- Jede Karte hat einen spezifischen Effekt
- Wichtiger strategischer Bestandteil für Spieleraktionen

### DreiZuEinsHafen
- Ermöglicht 3:1-Tausch beliebiger Ressourcen
- Befindet sich auf festgelegten Spielfeldplätzen
- Unterstützt Spieler beim Ressourcenmanagement

### Ecke
- Punkt auf dem Spielfeld für Bauwerke (Siedlungen/Städte)
- Bestimmt, welche Rohstoffe Spieler erhalten
- Kann frei oder durch ein Bauwerk besetzt sein

### Farbe
- Darstellung der Spielerkontrolle (Rot, Blau, Grün, Gelb)
- Identifikation von Spielern und Bauwerken
- Vermeidung von Verwechslungen im Spielfluss

### Hafen
- Bietet Handelsvorteile mit spezifischen Tauschverhältnissen
- Unterstützt den Spielerfortschritt durch effizienteren Rohstofftausch
- Verschiedene Häfen für verschiedene Ressourcen oder Tauschverhältnisse

### Kante
- Grenze zwischen zwei Ecken, für Straßenbau nutzbar
- Definiert den Expansionspfad der Spieler
- Verbindungselement zwischen Siedlungen/Städten

### Karte
- Oberbegriff für alle Spielerkarten (Aktions-, Ressourcen-, Sonderkarten)
- Repräsentiert Ressourcen oder Aktionen
- Ermöglicht Handels- und Bauaktivitäten

### Kartentyp
- Klassifizierung für Karten (Aktion, Ressource, Sonder)
- Hilft bei Strukturierung und Einsatz im Spiel
- Legt Karteneffekte und Verwendungen fest

### Räuber
- Spielfigur, blockiert Rohstofffelder temporär
- Kann Ressourcen von Mitspielern stehlen
- Beeinflusst Ressourcenfluss und Interaktion

### Regelwerk
- Definiert die Spielregeln
- Validiert Aktionen und reagiert auf Verstöße
- Zentrale Komponente für fairen Spielverlauf

### Ressource
- Stellt Rohstoffe für Bauwerke dar (Holz, Lehm, Weizen, Erz, Schaf)
- Grundlage für den Spielfortschritt jedes Spielers
- Wesentlich für Bau und Handel

### Ressourcenverwaltung
- Verteilung und Verwaltung der Rohstoffe
- Ermöglicht fairen Rohstofffluss zwischen Bank und Spielern
- Überwacht Ressourcenverfügbarkeit im Spiel

### Rohstofffeld
- Hexagon auf dem Spielfeld, das eine bestimmte Ressource produziert
- Produziert Rohstoffe bei Würfelzahlen
- Beeinflusst Verfügbarkeit von Ressourcen im Spiel

### Rohstoffkarte
- Stellt spezifische Rohstoffe für Spieler dar
- Wird für Bau und Handel genutzt
- Kern des wirtschaftlichen Spiels

### Runde
- Ein Zug eines Spielers, mit Aktionen wie Würfeln, Handeln, Bauen
- Strukturierter Spielfluss
- Sicherstellt, dass jeder Spieler nacheinander handelt

### Siedlung
- Bauwerk auf einer Ecke, das Ressourcen produziert
- Grundbaustein für den Aufbau einer Stadt
- Generiert Rohstoffe bei benachbarten Produktionsfeldern

### Sonderkarte
- Gewährt dem Besitzer besondere Boni
- Limitiert und strategisch wertvoll
- Kann den Spielverlauf entscheidend beeinflussen

### Spiel
- Repräsentiert den gesamten Spielablauf
- Enthält alle Komponenten wie Spieler, Spielbrett und Regeln
- Verwaltet Spielfortschritt und Aktionen

### Spielbrett
- Grundlage des Spiels, bestehend aus Feldern, Ecken, Kanten
- Ermöglicht Ressourcenproduktion und Handelsmöglichkeiten
- Zentrum für Spieleraktionen und strategische Positionierung

### Spieler
- Repräsentiert einen Siedler mit Farbe, Handkarten und Bauwerken
- Handelt und interagiert mit anderen Spielern
- Ziel: Höchste Punktzahl durch Bauwerke und Ressourcen sammeln

### Spielfigur
- Physische Elemente wie Straßen, Siedlungen, Städte
- Zeigen Besitz und Fortschritt des Spielers an
- Werden auf dem Spielbrett platziert und beeinflussen Ressourcenfluss

### Stadt
- Erweiterung einer Siedlung, erhöht Rohstoffproduktion
- Höchstes Bauwerk im Spiel
- Erhöht Siegpunkte und strategische Vorteile

### Straße
- Verbindung zwischen zwei Ecken für Expansion und Bau
- Ermöglicht Zugang zu neuen Ressourcen und Standorten
- Beeinflusst Expansionsmöglichkeiten des Spielers

### ZweiZuEinsHafen
- Ermöglicht 2:1-Tausch für spezifische Ressourcen
- Vorteil für Spieler, die sich auf bestimmte Ressourcen spezialisieren
- Strategische Handelsmöglichkeit zur Verbesserung des Rohstoffflusses
